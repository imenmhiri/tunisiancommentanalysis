 <a name="Data"><font color='#AA336A' style="font-size: 20px;"><strong>4- Data preparation </strong></font></a>

<font color='#A95C68' style="font-size: 15px;"><strong>  4.1 Data cleaning  </strong></font>

```python
features=df.iloc[:,0]
labels=df.iloc[:,1]
```

```python
emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0" # Miscellaneous symbols (checkmarks, musical notes, etc.)
                           u"\U000024C2-\U0001F251" # More symbols and pictographs (including arrows, geometric shapes, etc.)
                           "]+", flags=re.UNICODE) #This optional flag specifies that the regular expression engine should interpret the pattern using Unicode characters. 
regex = r"[^\u0600-\u06FF]"
tidy_features=[]
tidy_labels=[]
for i in range(len(features)):
    text=emoji_pattern.sub(r'', str(features[i])) # to remove emoji
    text=re.sub(r'https?://\S+|www\.\S+', ' ', text) # to remove link 
    text=re.sub(r"<.*?>", " ", text) #to remove html 
    text=re.sub(r'\d+', ' ', text) # to remove number
    text=re.sub(regex, " ", text) # to identify characters that are not Arabic.
    if text.strip():  # Check if text is not empty after stripping whitespace
        tidy_features.append(text)
        tidy_labels.append(labels[i])


df_cleaned = pd.DataFrame(data={'Cleaned_Text': tidy_features, 'labels': tidy_labels})
# to verify that uncessary charchter are removed from text 
df_cleaned.to_excel("cleaned_features.xlsx", index=False)

print("Preprocessing complete! Results saved to cleaned_features.xlsx")


```

```python
df_cleaned.info()
```

```python
# After cleaning the data by removing unnecessary characters, we identified a column containing null values so we need it
# to remove it 
df_result=df_cleaned.dropna()
df_result.info()
```

<font color='#A95C68' style="font-size: 15px;"> <strong>4.2 Data Normalisation  </strong></font>

```python
from pyarabic import araby
unified_text=[]
for i in range(len(tidy_features)):
    text = re.sub("[ÿ•ÿ£ÿ¢ÿß]", "ÿß", str(tidy_features[i]))
    text = re.sub("Ÿâ", "Ÿä", text)
    text = re.sub("ÿ§", "ÿ°", text)
    text = re.sub("ÿ¶", "ÿ°", text)
    text = re.sub("⁄Ø", "ŸÉ", text)
    # to remove ("ÿ∂ŸÖÿ© ŸÉÿ≥ÿ±ÿ© ŸÅÿ™ÿ≠ÿ© " "ÿ•ÿ¥ŸÉÿßŸÑ ") generally 
    text = araby.strip_diacritics(text)
    text = araby.strip_tatweel(text)
    unified_text.append(text)

# Create DataFrame and save to Excel
normalised_df = pd.DataFrame(data={'normalised_Text': unified_text, 'labels': tidy_labels})
normalised_df.to_excel("normalised.xlsx", index=False)



print("Preprocessing complete! Results saved to normalised.xlsx")
```

<div  style="font-family: Arial;box-sizing: border-box;
width: 75%;
border: solid #AA336A 3px;padding: 5px;margin: 0 auto;"><div style="margin-left: 10px ;margin-top: 15px;"><p style="font-size: 15px;"font color='#AA336A'><em><strong>fit :</strong></em> analyzes the text data in unified_text to learn the vocabulary (unique words) and calculate term frequencies (how often each word appears).</em></div>
<div style="margin-left: 10px ;;margin-top: 10px;"><p style="font-size: 15px;"font color='#AA336A'>
    <em> <strong>Transform : </strong></em> It converts each text document in unified_text into a TF-IDF vector representation. TF-IDF considers both the word frequency within a document and its overall frequency across the entire corpus, giving more weight to words that are important within a document but less frequent overall.</div></div>

```python
#all_stemmed_words = [word for sublist in stemmed_tokens for word in sublist]

vectorizer = TfidfVectorizer(stop_words=stopwords.words('arabic'))

X = vectorizer.fit_transform(unified_text).toarray()
```

<font color='#A95C68' style="font-size: 15px;"> <strong> 4.3 Word emedding usinf tf-idf Vectoriser </strong></font>

```python
X_train,X_test,y_train,y_test=train_test_split(X,tidy_labels,test_size=0.2)
```

<a name="Machine"><font color='#AA336A' style="font-size: 20px;"><strong> 5- Machine learning </strong></font></a>

```python
random_state = 40
```

```python

# the use of this MLP algorithms just to show that have bad performance in our case 
neural= MLPClassifier(hidden_layer_sizes=(100,150),activation='logistic',solver='adam',random_state=random_state)
gnb= GaussianNB()
#This strategy aims to compensate for the class imbalance during SVM training
linear_svm=SVC(kernel='linear',class_weight='balanced',random_state=random_state)
rbf_svm=SVC(kernel='rbf',class_weight='balanced',random_state=random_state)
sgd_svm=SVC(kernel='sigmoid',class_weight='balanced',random_state=random_state)
poly_svm=SVC(kernel='poly',degree=2,class_weight='balanced',random_state=random_state)

#training Ml algorithms
neural.fit(X_train,y_train)
gnb.fit(X_train,y_train)
linear_svm.fit(X_train,y_train)
rbf_svm.fit(X_train,y_train)
sgd_svm.fit(X_train,y_train)
poly_svm.fit(X_train,y_train)





```

```python
#Predection
y_pred_neural=neural.predict(X_test)
y_pred_linear=linear_svm.predict(X_test)
y_pred_rbf=rbf_svm.predict(X_test)
y_pred_sgd=sgd_svm.predict(X_test)
y_pred_poly=poly_svm.predict(X_test)
y_pred_gnd=gnb.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier


lass_weight = 'balanced'  
rf_model = RandomForestClassifier(n_estimators=100, random_state=random_state, class_weight=lass_weight,criterion='entropy')

#training Ml algorithms
rf_model.fit(X_train, y_train)
```

```python
#Predection
y_pred_forest = rf_model.predict(X_test)
```

<a name="Performance"><font color='#AA336A' style="font-size: 20px;"><strong> 6- Performance evaluation  </strong></font></a>

```python
print("************* performance evaluation of Naive bayes ***********")
print(classification_report(y_test,y_pred_linear))
print("************* performance evaluation of MLP Classifier  ***********")
print(classification_report(y_test,y_pred_neural))
print("************* performance evaluation of RBF SVM ***********")
print(classification_report(y_test,y_pred_rbf))
print("************* performance evaluation of SGD SVM ***********")
print(classification_report(y_test,y_pred_sgd))
print("************* performance evaluation of POLYNOMAIL SVM ***********")
print(classification_report(y_test,y_pred_poly))
print("************* performance evaluation of linear SVM ***********")
print(classification_report(y_test,y_pred_linear))
print("************* performance evaluation of random forest Classifier ***********")
print(classification_report(y_test,y_pred_forest))


```
